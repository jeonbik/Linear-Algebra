```python
alpha = np.linspace(0, 2 * np.pi, 50)
vectors = list(zip(np.cos(alpha),np.sin(alpha)))
plot_vector(vectors)
```

```python
new_vectors = []
for i in range(len(vectors)):
    new_vectors.append(A.dot(np.array(vectors[i])))
plot_vector(new_vectors)
```
