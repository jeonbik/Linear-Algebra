```python
a = np.array((-2,1))
b = np.array((3,6))
origin = np.array((0,0))

plot_vector([a], [origin])
plt.title("Vector A")
plt.show()

plot_vector([b], [origin])
plt.title("Vector B")
plt.show()

vector = [a,b, a+b]
tails = [origin, a, origin]

plot_vector(vector,tails)
plt.title("Vector Addition: A + B")
plt.show()
```
